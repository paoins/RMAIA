```python
import yfinance as yf
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
# SQLAlchemy engine
engine = create_engine(f"sqlite:///rmaia_market_data.db")
print('connection created')
```

```python
try:
  # Download Data
  data = yf.download(
      tickers = ['BTC-USD', 'EURUSD=X', 'GLD'],
      period = '5y',
      interval = '1d'
  )

  # Flatten the MultiIndex
  data.columns = ['_'.join(col).strip() for col in data.columns.values]
  data.index.name = 'Date'

  # Write to SQlite database
  data.to_sql(
      name = 'daily_historical',
      con = engine,
      if_exists = 'replace',
      index = True
  )
  print('data written to database')

  # Read the Data to comfirm
  df = pd.read_sql(
      f"SELECT * FROM {'daily_historical'} LIMIT 5", con=engine)
  print(df)
except Exception as e:
  print(e)
```
